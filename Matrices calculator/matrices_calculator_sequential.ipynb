```python
import numpy as np
import timeit
```

```python
start = timeit.default_timer()
```

```python
def mult_matrices(X , Y , result):
    # iterate through rows of X
    for i in range(len(X)):
       # iterate through columns of Y
       for j in range(len(Y[0])):
           # iterate through rows of Y
           for k in range(len(Y)):
               result[i][j] += X[i][k] * Y[k][j]
    return result
```

```python
def add_matrices(X , Y , result):
    # iterate through rows
    for i in range(len(X)):
       # iterate through columns
       for j in range(len(X[0])):
           result[i][j] = X[i][j] + Y[i][j]
    return result
```

```python
X = np.arange(0, 600).reshape(20,30)

Y = np.arange(0, 600).reshape(20,30)

Z = np.arange(0, 600).reshape(20,30)

N = np.arange(0, 600).reshape(20,30)

result = np.zeros((20,30))
```

```python
f1 = add_matrices(X, Y , result)
f2 = add_matrices(Z, N , result)
```

```python
print("Result of Addition (X , Y):")
for a in f1:
   print(a)
```

```python
print("Result of Addition (Z , N):")
for a in f2:
   print(a)
```

```python
mult_result = mult_matrices(f1, f2 , result)

print("Result of multiplication:")
for m in mult_result:
   print(m)
```

```python
stop = timeit.default_timer()

print('Time: ', stop - start) 
```
